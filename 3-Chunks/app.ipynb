# Dividing Data info chunks

#### - Chunking is dividing text data into pieces for text analysis
#### - it's not the same as Tokenization, chunks need to be meaningful
#### - Large text -> Devide text to chunks

```python
# import text example
import nltk
from nltk.corpus import brown
```

```python
nltk.download('brown')
```

```python
# Split the input text into chunks, where
# each chunk contains N words
def chunker(input_data, N):
    input_words =input_data.split(' ')
    output =[]
    
    curr_chunk =[]
    count =0
    
    for word in input_words:
        curr_chunk.append(word)
        count+=1
        if count == N:
            output.append(' '.join(curr_chunk))
            count, curr_chunk =0, []
            
    output.append(' '.join(curr_chunk))
    return output
```

```python
input_data = ' '.join(brown.words())
chunk_size =700
```

```python
chunks =chunker(input_data, chunk_size)
```

```python
print('\nNumber of text chunks = ', len(chunks),'\n')

for i, chunk in enumerate(chunks):
    print('Chunk', i+1, '===>', chunk[:50])
```
